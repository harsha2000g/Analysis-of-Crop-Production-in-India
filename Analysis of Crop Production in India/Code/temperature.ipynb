```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('temperatures.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
fig, axs = plt.subplots(ncols=2,figsize=(20,10))
sns.scatterplot(x=dataset['YEAR'], y=dataset['JAN-FEB'],data=dataset, ax=axs[0])
sns.scatterplot(x=dataset['YEAR'], y=dataset['MAR-MAY'],data=dataset, ax=axs[1])
fig, axs = plt.subplots(ncols=2,figsize=(20,10))
sns.scatterplot(x=dataset['YEAR'], y=dataset['JUN-SEP'],data=dataset, ax=axs[0])
sns.scatterplot(x=dataset['YEAR'], y=dataset['OCT-DEC'],data=dataset, ax=axs[1])
fig, axs = plt.subplots(ncols=1,figsize=(20,10))
sns.scatterplot(x=dataset['YEAR'], y=dataset['ANNUAL'], data=dataset)
```

```python
fig, axs = plt.subplots(ncols=5,figsize=(20,10))
sns.boxplot(dataset['ANNUAL'],orient="v", data=dataset, ax=axs[0])
sns.boxplot(dataset['JAN-FEB'],orient="v", data=dataset, ax=axs[1])
sns.boxplot(dataset['MAR-MAY'],orient="v", data=dataset, ax=axs[2])
sns.boxplot(dataset['JUN-SEP'],orient="v", data=dataset, ax=axs[3])
sns.boxplot(dataset['OCT-DEC'],orient="v", data=dataset, ax=axs[4])
```

```python
corr = dataset.corr()
plt.figure(num=None, figsize=(10,10), dpi=80, facecolor='w', edgecolor='k')
corrMat = plt.matshow(corr, fignum = 1)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.gca().xaxis.tick_bottom()
plt.colorbar(corrMat)
plt.title('Correlation Matrix', fontsize=15)
plt.show()
```
